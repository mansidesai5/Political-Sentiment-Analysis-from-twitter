```python
from textblob import TextBlob

modi_txt=TextBlob(Modi["Tweet"][2])
print(Modi["Tweet"][2])
print("Narendra Modi:", modi_txt.sentiment)

rgandhi_txt=TextBlob(RGandhi["Tweet"][10])
print(RGandhi["Tweet"][10])
print("Rahul Gandhi:", rgandhi_txt.sentiment)
```

```python
def pol(Tweet) :
  return TextBlob(Tweet).sentiment.polarity

Modi["Polarity"] = Modi["Tweet"].apply(pol)
Modi.head(20)
```

```python
RGandhi["Polarity"] = RGandhi["Tweet"].apply(pol)
RGandhi.head(20)
```

```python
import numpy as np

Modi["sentiment"] = np.where(Modi["Polarity"]>0,"Positive","Negative")
Modi["sentiment"][Modi["Polarity"]==0]="Neutral"
Modi.head(20)
```

```python
RGandhi["sentiment"] = np.where(RGandhi["Polarity"]>0,"Positive","Negative")
RGandhi["sentiment"][RGandhi["Polarity"]==0]="Neutral"
RGandhi.head(20)
```
